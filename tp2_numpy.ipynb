IV.Conversion dâ€™une image couleur en image en niveau de gris

Question 1

```python
def NiveauGris(img):
  L = img.shape[0]
  H = img.shape[1]
  img_grey = np.zeros((L,H))
  for i in range(L):
      for j in range(H): 
        img_grey[i,j] = ( 0.299*img[i,j,0] +  0.587*img[i,j,1] +  0.114*img[i,j,2] )
  plt.imshow(img_grey, cmap = 'Greys_r')
```
